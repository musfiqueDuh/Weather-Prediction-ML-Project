```python
# Define the constants for the formula -Feature Engineering
c1 = -8.78469475556
c2 = 1.61139411
c3 = 2.33854883889
c4 = -0.14611605
c5 = -0.012308094
c6 = -0.0164248277778
c7 = 0.002211732
c8 = 0.00072546
c9 = -0.000003582

# Calculate heat index using the formula
def calculate_heat_index(temperature_celsius, humidity):
    heat_index = c1 + (c2 * temperature_celsius) + (c3 * humidity) + (c4 * temperature_celsius * humidity) + (c5 * temperature_celsius**2) + (c6 * humidity**2) + (c7 * temperature_celsius**2 * humidity) + (c8 * temperature_celsius * humidity**2) + (c9 * temperature_celsius**2 * humidity**2)
    return heat_index

# Add heat index as a new feature
weather_data['heat_index'] = calculate_heat_index(weather_data['temperature_celsius'], weather_data['humidity'])

# Check the updated dataset
print(weather_data[['temperature_celsius', 'humidity', 'heat_index']].head())

```

```python
from sklearn.preprocessing import LabelEncoder

# Example: Encoding categorical variables
label_encoder = LabelEncoder()
weather_data['country label'] = label_encoder.fit_transform(weather_data['country'])
# Repeat this for other categorical variables if present

weather_data
```

```python
weather_data['timezone label'] = label_encoder.fit_transform(weather_data['timezone'])
weather_data['location_name label'] = label_encoder.fit_transform(weather_data['location_name'])
weather_data['condition_text label'] = label_encoder.fit_transform(weather_data['condition_text'])
weather_data['sunrise label'] = label_encoder.fit_transform(weather_data['sunrise'])
weather_data['sunset label'] = label_encoder.fit_transform(weather_data['sunset'])                
weather_data['moonrise label'] = label_encoder.fit_transform(weather_data['moonrise'])
weather_data['moonset label'] = label_encoder.fit_transform(weather_data['moonset'])
weather_data['moon_phase label'] = label_encoder.fit_transform(weather_data['moon_phase'])
weather_data['wind_direction label'] = label_encoder.fit_transform(weather_data['wind_direction'])

weather_data
```

```python
weather_data = weather_data.drop(['country','location_name','timezone','condition_text','sunrise','sunset','moonrise','moonset','wind_direction','moon_phase'],axis=1)
weather_data.head()
```

```python

import seaborn as sns

# Plotting the distribution of heat index categories
plt.figure(figsize=(8, 6))
sns.countplot(x='heat_index_category', data=weather_data, palette='coolwarm', order=['Low Heat', 'Normal Heat', 'Excess Heat'])
plt.title('Distribution of Heat Index Categories')
plt.xlabel('Heat Index Category')
plt.ylabel('Count')
plt.show()
```

```python
# Calculating mean cloud cover for each heat index category
mean_cloud_by_category = weather_data.groupby('heat_index_category')['wind_direction label'].mean()

# Bar chart of average cloud cover by heat index category
plt.figure(figsize=(10, 6))
mean_cloud_by_category.plot(kind='bar', color=['blue', 'grey', 'red'])
plt.title('Average Wind Direction by Heat Index Category')
plt.xlabel('Heat Index Category')
plt.ylabel('Wind Direction')
plt.xticks(rotation=0)  # Ensure category names are horizontal
plt.show()
```
